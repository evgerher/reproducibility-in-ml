```python
import pandas as pd
```

```python
train = pd.read_csv('data/train.csv')
test = pd.read_csv('data/test.csv')
```

```python
season_mapping = {1: 'Spring', 
                  2: 'Summer', 
                  3: 'Fall', 
                  4: 'Winter'}
weather_mapping = {1: 'Clear', 
                   2: 'Mist, Few clouds', 
                   3: 'Light Snow, Rain, Thunder',
                   4: 'Heavy Snow, Rain, Thunder'}
weekday_mapping = {0: 'Monday', 1: 'Tuesday', 2: 'Wednesday', 
                   3: 'Thursday', 4: 'Friday', 5: 'Saturday', 6: 'Sunday'}
workingday_mapping = {0: 'Working day', 1: 'Weekend or Holiday'}
holiday_mapping = {0: 'Working day', 1: 'Holiday'}

train
```

```python
train['holiday'] = train['holiday'].map(holiday_mapping)
train['workingday'] = train['workingday'].map(workingday_mapping)
train['season'] = train['season'].map(season_mapping)
train['weather'] = train['weather'].map(weather_mapping)

test['holiday'] = test['holiday'].map(holiday_mapping)
test['workingday'] = test['workingday'].map(workingday_mapping)
test['season'] = test['season'].map(season_mapping)
test['weather'] = test['weather'].map(weather_mapping)

# train['weekday'] = train['weekday'].map(holiday_mapping)
```
